<h4 style = "text-align:left;">a) Import key libraries</h4>

```python
import seaborn as sns
import pandas as pd
```

<h4 style = "text-align:left;">b) Load Data (use seaborn built-in data) </h4>

```python
# Load seaborn data for tips, iris, titanic and planets 
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
print("First 5 Rows of Tips Dataset")
print("-"*30)
print(tips.head())
print("\n")
print("First 5 Rows of Iris Dataset")
print("-"*30)
print(iris.head())
print("\n")
print("First 5 Rows of Titatic Dataset")
print("-"*32)
print(titanic.head())
print("\n")
print("First 5 Rows of Planets Dataset")
print("-"*32)
print(planets.head())
```

<h4 style = "text-align:left;">c) Seaborn Scatter Plots </h4>

```python
sns.scatterplot(x = "tip", 
                y = "total_bill", 
                data = tips, 
                hue = "day", 
                size = "size",
                palette = 'YlGnBu')
```

<h4 style = "text-align:lefft;">d) Seaborn Histogram</h4>

```python
sns.histplot(tips['tip'],
             kde = True, 
             bins = 20)
```

```python
sns.displot(tips["tip"],
            kde = False, 
            bins = 50)
```

<h4 style = "text-align:left;">e) Seaborn Bar Chart</h4>

```python
sns.barplot(x = "sex",
            y = "tip",
            data = tips,
            palette = "YlGnBu" )
```

<h4 style = "text-align:left;">f) Seaborn Box Plot</h4>

```python
sns.boxplot(x = "day",
            y = "tip", 
            data = tips, 
            hue = "sex", 
            palette = "YlGnBu")
```

```python
sns.boxplot(x = "day",
            y = "total_bill",
            data = tips,
            hue = "sex", 
            palette = "YlGnBu")
```

<h4 style = "text-align:left;">g) Seaborn Strip Plot</h4>

```python
sns.stripplot(x = "day", 
              y = "tip",
              data = tips,
              hue = "sex",
              palette = "YlGnBu")
```

```python
sns.stripplot(x = "day",
              y = "tip",
              data = tips,
              hue = "sex",
              palette = "YlGnBu", 
              dodge = True)
```

<h4 style = "text-align:left;">h) Seaborn Joint Plot</h4>

```python
sns.jointplot(x = "tip",
              y = "total_bill", 
              data = tips,
              hue = "sex")
```

```python
sns.jointplot(x = "tip", 
              y = "total_bill", 
              data = tips, 
              kind = "reg")
```

```python
sns.jointplot(x = "tip", 
              y = "total_bill", 
              data = tips, 
              kind = "kde",
              shade = True,
              cmap = "YlGnBu")
```

<h4 style = "text-align: leftt;">i) Seaborn Pair Plot</h4>

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.pairplot(titanic.select_dtypes(["number"]), hue = "pclass")
```

<h4 style = "text-align:left;">j) Seaborn Heatmap for correlation</h4>

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "YlGnBu")
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "coolwarm")
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "icefire")
```
