```python
# Import libraries 
import matplotlib.pyplot as plt
#%matplotlib inline
#plt.style.use('fivethirtyeight')
#plt.style.use("dark_background")
#plt.style.use("ggplot")
import numpy as np
import pandas as pd 
```

```python
# Create imaginary dataset
coffee_sales = [20000, 18000, 15000, 3500, 16500, 8000, 4500, 17500, 3700, 8600, 5000]
temperature = [10, 14, 23, 39, 18, 30, 35, 21, 37, 27, 33]
```

```python
#Method 1: Scatter plot: 
plt.figure(figsize = (10, 5))
plt.scatter(coffee_sales, temperature, c = "green", marker = "*" )

# Change axes ranges 
plt.xlim(0, 22000)
plt.ylim(0, 50)

# Add title and set fontsize 
plt.title("Relationship Between Temperature and Latte Coffee Sales", fontsize = 10)

#Add x and y labels
plt.xlabel("Coffee Sales (Kenya Shillings)", fontsize = 9)
plt.ylabel("Temperature (degree)", fontsize = 9)

#Show plot
plt.show()
```

```python
# Method 2: Scatter plot:
# Set the size of the plot
plt.figure(figsize = (10, 5))
plt.scatter(coffee_sales, temperature, c = "red", marker = "+" )

# Add title and set fontsize 
plt.title("Relationship Between Temperature and Latte Coffee Sales",
           fontsize = 12,
           fontname = "serif")

#Add x and y labels
plt.xlabel("Coffee Sales (Kenya Shillings)", 
           fontsize = 10,
           fontname = "cursive" )
plt.ylabel("Temperature (degree)",
           fontsize = 10, 
           fontname = "cursive")

#Show plot
plt.show()
```

```python
# Create an empty figure 
fig = plt.figure(figsize = (10, 6))
ax = plt.axes()

# Generate data
x = np.linspace(start = 0, stop = 10, num = 1000)

# Plot x and sin(x)
ax.plot(x, np.sin(x))

# Plot x and cos(x)
ax.plot(x, np.cos(x))

plt.axis("tight")
# Add title 
plt.title("Graph x and sin (x) and cos (x)", fontsize = 14)

# Add both x and y lables 
plt.xlabel("values of x", fontsize = 12)
plt.ylabel("sin (x) and cos (x)", fontsize = 12)

plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))
ax = plt.axes()

x = np.linspace(start=0, stop=10, num=1000)

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.plot(x, (np.sin(x) + np.cos(x)))
plt.show()
```

```python
plt.figure(figsize = (10, 5))
ax = plt.axes()
x = np.linspace(start=0, stop=10, num=1000)
ax.plot(x, np.sin(x - 0), color = "blue")             # specifies colour by name 
ax.plot(x, np.sin(x - 1), color = "g")                # shortcut for color code (rgbcmyk)
ax.plot(x, np.sin(x - 2), color = "0.75")             # greyscale between 0 to 1
ax.plot(x, np.sin(x - 3), color = "#FFDD44")          # Hex code (RRGGBB for OO to FF)
ax.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))    # RGB tuple, values from 0 to 1
ax.plot(x, np.sin(x - 5), color = "chartreuse")       # all HTML colors supported 
```

```python
plt.figure(figsize = (10, 5))
ax = plt.axes()
x = np.linspace(start=0, stop=10, num=1000)
ax.plot(x, np.cos(x - 0), color = "blue")             # specifies colour by name 
ax.plot(x, np.cos(x - 1), color = "g")                # shortcut for color code (rgbcmyk)
ax.plot(x, np.cos(x - 2), color = "0.75")             # greyscale between 0 to 1
ax.plot(x, np.cos(x - 3), color = "#FFDD44")          # Hex code (RRGGBB for OO to FF)
ax.plot(x, np.cos(x - 4), color = (1.0, 0.2, 0.3))    # RGB tuple, values from 0 to 1
ax.plot(x, np.cos(x - 5), color = "chartreuse")       # all HTML colors supported 
```

```python
# Create some data 
data = np.random.normal(loc = 0, scale = 1.5, size = 1000) # loc = mean, scale = std 
plt.figure(figsize = (10, 5))
plt.plot(data, color = "r")
plt.title("Normally Distributed Data", fontsize = 12)
plt.show()
```

```python
# Histogram

data = np.random.normal(loc = 0, scale = 1.5, size = 1000) # loc = mean, scale = std 
plt.figure(figsize = (10, 5))
plt.hist(data,bins = 50, color = "b")
plt.title("Normally Distributed Data", fontsize = 12)
plt.show()
```

```python
# Subplots 
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
plt.figure(figsize = (5,5))
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
# Share a X axis with each column of subplots
fig, axs = plt.subplots(2, 2, sharex='col')
x = np.linspace(start = 0, stop = 10, num = 1000)
axs[0,0].scatter(x, np.sin(x), color = "red")
axs[0,1].plot(x, np.cos(x**2), color = "g")
axs[1,0].plot(x, np.cos(x), color = "0.05")
axs[1,1].plot(x, np.tan(x))
```

```python
# 3 Dimensional Plots 
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random (100)

ax = plt.axes(projection = "3d")
ax.scatter(x, y, z)

# Set title of the plot
ax.set_title("3 Dimensional Plot", fontsize = 14)

# Add labels to the axes 
ax.set_xlabel("test")

plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(start =0, stop = 50, step = 0.2)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title("3 Dimensional Plot for X, sin (x) and cos (x)")
ax.set_xlabel("test")
plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(0, 50, 0.1)
y = np.arange(0, 50, 0.1)
z = np.cos(x + y)

ax.plot(x, y, z)
```

```python
ax = plt.axes(projection = "3d")
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X, Y, Z, cmap = "Spectral")
plt.show()
```
