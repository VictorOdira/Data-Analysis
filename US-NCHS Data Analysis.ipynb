<h4 style = "text-align:left;">1. Import relevant libraries</h4>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

<h4 style = "text-align:left;">2. Load CSV data from the US National Center for Health Statistcs</h4>

```python
data = pd.read_csv("NCHS.csv")
data
```

<h4 style = "text-align:left;">3. How may rows and columns are in the dataset?</h4>

```python
# Method 1
print(f"Dataset has {data.shape[0]} rows and {data.shape[1]} columns")

# Method 2
print("Dataset has {} rows and {} columns".format(data.shape[0], data.shape[1]))
```

<h4 style = "text-align:left;">4. Ascertain if the dataset has missing values (nan) </h4>

```python
data.isna().sum()
```

```python
data.isnull().any()
```

<h4 style = "text-align:left;">5. Remove all rows with na cases if any</h4>

```python
data.dropna()
data.shape
```

<h4 style = "text-align:left;">6. What are the unique causes of death in the dataset?</h4>

```python
causes_of_death = data["Cause Name"].unique()
print(causes_of_death)
```

<h4 style = "text-align:left;">7. Remove "All causes" from the Cause Name Column</h4>

```python
data = data[data["Cause Name"] != "All causes"]
causes_of_death = data["Cause Name"].unique()
print(causes_of_death)
```

<h4 style = "text-align:left;">8. How many causes of death are in the dataset?</h4>

```python
number_of_causes = len(causes_of_death)
print(f"Dataset has {number_of_causes} causes of death")
```

<h4 style = "text-align:left;">9. What are the States which are affected by the causes of death?</h4>

```python
affected_states = data["State"].unique()
print(affected_states)
```

<h4 style = "text-align:left;">10. Remove "United States" from the "State" Column</h4>

```python
data = data[data["State"] != "United States"]
affected_states = data["State"].unique()
print(affected_states)
```

<h4 style = "text-align:left;">11. How many states are affected?</h4>

```python
number_of_states = len(affected_states)
print("There are %s states affected in the dataset" %number_of_states)
```

<h4 style = "text-align:left;">12. What are the total 
number of deaths in the United States from 1999 to 2016?</h4>

```python
data.head(0)
```

```python
total_deaths = data["Deaths"].sum()
print("%s deaths were registered between 1999 and 2016"%total_deaths)
```

<h4 style = "text-align:left;">13. What are the number of deaths per year?</h4>

```python
deaths_per_year = data.groupby("Year").sum()
print(deaths_per_year)
```

<h4 style = "text-align:left;">14. What was trend of deaths per year?</h4>

```python
plt.figure()
deaths_per_year["Deaths"].plot(color = "g")
plt.title("Yearly deaths from 1999 to 2016", fontsize = 14, fontname = "serif")
plt.xlabel("Year", fontsize = 10, fontname="serif")
plt.ylabel("Yearly deaths (in millions)", fontsize = 10, fontname = "serif")
plt.show()
```

<h4 style = "text-align:left;">15. Which 10 states had the highest number of deaths in all years?</h4>

```python
data_df = data[data["State"] != "United States"]
dataset = data_df.groupby("State").sum()
dataset.sort_values("Deaths", ascending = False, inplace = True)
top_10 = dataset.head(10)
print(top_10)
```

```python
top_10["Deaths"].plot.bar()
plt.title("Top 10 States with the highest number of deaths", fontsize = 12, fontname = "fantasy")
plt.xlabel("State", fontsize = 10, fontname = "serif")
plt.ylabel("Number of deaths ('000,000)", fontsize = 10, fontname = "serif")
plt.show()
```

<h4 style = "text-align:left;">15. What were the 10 States the least number of deaths between 1999 and 2016?</h4>

```python
data_df = data[data["State"] != "United States"]
dataset = data_df.groupby("State").sum()
dataset.sort_values("Deaths", ascending = False, inplace = True)
last_10 = dataset.tail(10)
print(last_10)
```

<h4 style = "text-align:left;">15. What were the top causes of deaths in the United States between 1999 and 2016?</h4>

```python
data_df = data[data["Cause Name"] != "All causes"]
dataset2 = data_df.groupby("Cause Name").sum()
dataset2.sort_values("Deaths", ascending = False, inplace = True)
top_10_causes = dataset2.head(10)
print(top_10_causes)
```

```python
top_10_causes["Deaths"].plot.pie(autopct = "%.1f%%")
```
